## ATM

Note that with `adjust_month=False`, the default output from CESM is offset by one month.

```python
dirpath = '/glade/campaign/univ/ubrn0018/fengzhu/CESM_output/timeseries/b.e13.B1850C5.ne16_g16.icesm131_d18O_fixer.Miocene.3xCO2.005/atm/proc/tseries/month_1'
fname = 'b.e13.B1850C5.ne16_g16.icesm131_d18O_fixer.Miocene.3xCO2.005.cam.h0.TS.695101-700012.nc'
ds = xpp.load_dataset(os.path.join(dirpath, fname), adjust_month=False)
ds
```

Therefore, we should always set `adjust_month=True` when dealing with CESM output.

```python
ds = xpp.load_dataset(os.path.join(dirpath, fname), adjust_month=True)
ds
```

By default, the `annualize()` method will take a calendar year annualization.
It works on both `xarray.Dataset` and `xarray.DataArray`.

```python
da_ann = ds.xpp.annualize()
da_ann
```

```python
da_ann = ds['TS'].xpp.annualize()
da_ann
```

Now let's take the summer annualization (JJA) with `months=[6, 7, 8]`.

```python
da_ann = ds['TS'].xpp.annualize(months=[6, 7, 8])
da_ann
```

Now let's take the winter annualization (DJF) with `months=[-12, 1, 2]`.
Note that for such cases with months acrossing two years, the output will have an extra year.

```python
da_ann = ds['TS'].xpp.annualize(months=[-12, 1, 2])
da_ann
```
